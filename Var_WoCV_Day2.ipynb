```python
#finding and drawing contours
img2 = cv2.imread("/content/yup.png")
img_gray = cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY)
blur = cv2.GaussianBlur(img_gray,(11,11
                                  ),3)
ret, thresh = cv2.threshold(blur, 235, 255, cv2.THRESH_BINARY)
contours, hierarchy = cv2.findContours(image=thresh, mode=cv2.RETR_TREE, method=cv2.CHAIN_APPROX_NONE)
image_copy = img2.copy()

cv2.drawContours(image_copy,contours, -1, (0, 255, 0),2, cv2.LINE_AA)
# see the results
cv2_imshow(image_copy)
cv2.imwrite('contours_none_image1.jpg', image_copy)

#finding centers of the contours
for i in contours:
    M = cv2.moments(i)
    if M['m00'] != 0:
        cx = int(M['m10']/M['m00'])
        cy = int(M['m01']/M['m00'])
        cv2.drawContours(img2, [i], -1, (0, 255, 0), 2)
        cv2.circle(img2, (cx, cy), 7, (0, 0, 255), -1)
        cv2.putText(img2, "center", (cx - 20, cy - 20),
                   cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 0, 0), 2)


cv2_imshow(img2)
```

```python

import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import numpy as np
import cv2

def Cartoon(image_color):
    output_image = cv2.stylization(image_color, sigma_s=100, sigma_r=0.3)
    return output_image

def LiveCamEdgeDetection_canny(image_color):
    threshold_1 = 30
    threshold_2 = 80
    image_gray = cv2.cvtColor(image_color, cv2.COLOR_BGR2GRAY)
    canny = cv2.Canny(image_gray, threshold_1, threshold_2)
    return canny
# Main calling function to initialize webcam and apply edge detection

cap = cv2.VideoCapture(0)

while True:
    ret, frame = cap.read()
    cv2_imshow(LiveCamEdgeDetection_canny(frame))
    cv2_imshow(frame)
    if cv2.waitKey(1) == 13:
        break

cap.release()
cv2.destroyAllWindows()
```

```python
image = cv2.imread('/content/cap.png')
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.imshow(gray, cmap='gray')

blur = cv2.GaussianBlur(gray, (11, 11), 0)
plt.imshow(blur, cmap='gray')

canny = cv2.Canny(blur, 30, 120, 3)
plt.imshow(canny, cmap='gray')

dilated = cv2.dilate(canny, (1, 1), iterations=0)
plt.imshow(dilated, cmap='gray')

(cnt, hierarchy) = cv2.findContours(
	dilated.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)
rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
cv2.drawContours(rgb, cnt, -1, (0, 220, 225), 2)

plt.imshow(rgb)

print("coins in the image : ", len(cnt))

#not accurate tho
```
